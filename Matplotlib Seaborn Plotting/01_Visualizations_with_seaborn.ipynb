```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
sns.distplot(df['petal_length'], kde=False)
plt.show()
```

```python
sns.distplot(df['petal_length'], hist=False)
plt.show()
```

```python
sns.distplot(df['petal_length'])
plt.show()
```

# Plotting bivariate distribution

```python
sns.jointplot(x = 'petal_length', y = 'petal_width', data=df)
plt.show()
```

```python
sns.jointplot(x = 'petal_length', y = 'petal_width', 
              data=df, kind='hex')
plt.show()
```

```python
sns.set_style("ticks")
sns.pairplot(df, hue="species", diag_kind="kde",
            kind="scatter", palette="husl")
plt.show()
```

```python
sns.stripplot(x="species", y="petal_length", data=df)
plt.show()
```

```python
sns.stripplot(x="species", y="petal_length", 
              data=df, jitter=True)
plt.show()
```

```python
sns.swarmplot(x="species", y="petal_length", data=df)
plt.show()
```

```python
df = sns.load_dataset('tips')
sns.violinplot(x='day', y = 'total_bill', data=df)
plt.show()
```

```python
sns.violinplot(x='day', y = 'total_bill', 
               hue='sex', data=df)
plt.show()
```

```python
df = sns.load_dataset('titanic')
sns.barplot(x='sex', y='survived',
           hue='class', data=df)
plt.show()
```

```python
sns.countplot(x='class', data=df, palette="Blues")
plt.show()
```

```python
sns.pointplot(x='sex', data=df, y="survived", hue='class')
plt.show()
```

```python
df = sns.load_dataset('iris')
sns.boxplot(data= df, orient='h')
plt.show()
```

```python
sns.boxplot(data= df, orient='v')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(5, 6))
df = sns.load_dataset('iris')
sns.regplot(x='sepal_length',
           y= 'petal_length',
           data=df, ax=ax)
plt.show()
```
