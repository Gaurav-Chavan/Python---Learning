# Summary Statistics by Level

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame(np.arange(12).reshape((4, 3)),
                  index=[['a', 'a', 'b', 'b'], [1, 2, 1, 2]],
                        columns = [['Mumbai', 'Mumbai', 'Delhi'],
                                  ['Red', 'Green', 'Blue']])
```

```python
df1.index.names = ['key1', 'key2']
```

```python
df1
```

```python
df1.sum(level='key1')
```

```python
df1.sum(level='key2')
```

```python
df1.min(level='key1')
```

```python
df1.max(level='key1')
```

# Indexing with a DataFrame's columns

```python
df2
```

```python
df3 = df2.set_index(['D', 'C'])
```
