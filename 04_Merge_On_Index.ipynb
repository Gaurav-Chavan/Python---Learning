```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                    'value': range(6)})
df2 = pd.DataFrame({'val': [1, 2]}, index=['a', 'b'])
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, left_on='key', right_index=True)
```

```python
pd.merge(df1, df2, left_on='key', right_index=True, how='outer')
```

```python
df1 = pd.DataFrame({'key1': ['Mumbai', 'Mumbai', 'Mumbai', 'Delhi', 'Delhi'],
                    'key2': [2000, 2001, 2002, 2001, 2002],
                    'data': np.arange(5)})

df2 = pd.DataFrame(np.arange(12).reshape((6, 2)),
                                index=[['Delhi', 'Delhi', 'Mumbai', 
                                        'Mumbai', 'Mumbai', 'Mumbai'],
                                      [2001, 2000, 2000, 2000, 2001, 2001]],
                                columns=['event1', 'event2'])
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, left_on=['key1', 'key2'], right_index=True)
```

```python
pd.merge(df1, df2, left_on=['key1', 'key2'], right_index=True, how = 'outer')
```

```python
df1 = pd.DataFrame([[1, 2], [3, 4], [5, 6]],
                   index = ['a', 'c', 'e'],
                   columns=['Mumbai', 'New Delhi'])
df2 = pd.DataFrame([[7, 8], [9, 10], [11, 12], [13, 14]],
                   index = ['b', 'c', 'd', 'e'],
                   columns=['Bangalore', 'Chennai'])
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, left_index=True, right_index=True, how='outer')
```

```python
pd.merge(df1, df2, left_index=True, right_index=True)
```

```python
df1.join(df2, how = 'outer')
```

```python
df1.join(df2)
```

```python
df1.join(df2, how = 'inner')
```
